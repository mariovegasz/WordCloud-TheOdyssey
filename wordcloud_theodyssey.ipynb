```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image # tratamiento de imagenes
import PIL
import nltk
nltk.download('stopwords')
```

```python
stopwords=set(nltk.corpus.stopwords.words("spanish"))
```

```python
odisea = open('odisea.txt', 'r',encoding="UTF-8").read()
```

```python
len(odisea)
```

```python
odisea_wc = WordCloud(background_color='white', max_words=len(odisea), stopwords=stopwords)
odisea_wc.generate(odisea)
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax.imshow(odisea_wc)
ax.axis('off')
plt.show()
```

```python
stopwords.add('si') 
stopwords.add('pues') 
stopwords.add('ahora') 
odisea_wc = WordCloud(background_color='white', max_words=len(odisea), stopwords=stopwords)
odisea_wc.generate(odisea)
fig, ax = plt.subplots(figsize=(12,12))
ax.imshow(odisea_wc)
ax.axis('off')
plt.show()
```

```python
mask = np.array(Image.open('lobo.jpg'))   
mask.shape
```

```python
img = mask.copy()
```

```python
plt.imshow(img)
```

```python
# Creamos de nuevo el objeto agregando la mascara
alice_wc = WordCloud(background_color="bisque", max_words=2000, mask=img,
               stopwords=stopwords, contour_width=0.5, contour_color='black',colormap = "plasma")

alice_wc.generate(odisea)


fig, ax = plt.subplots(figsize=(15,15))
ax.imshow(alice_wc)
ax.axis('off')
alice_wc.to_file("lobo.jpg")
plt.show()
```
